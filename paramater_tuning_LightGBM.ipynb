```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML 
import seaborn as sns
import json
import ast
```

```python
df = pd.read_csv('param_tuning_lightgbm.csv')
```

```python
df.sort_values('loss')
```

```python
accuracy = (1-df['loss'])
```

```python
accuracy = accuracy.rename('acc')
```

```python
new_df = pd.concat([df, accuracy],axis=1)
```

```python
new_df
```

```python
new_df.sort_values('acc', ascending=False)
```

```python
plt.ylim(0.74, 0.79)
plt.scatter(new_df['iteration'], new_df['acc'])
```

```python
imsize = 20
plt.figure(figsize=(imsize,imsize))
sns.jointplot(new_df['iteration'], new_df['acc'], kind='reg')
```

```python
best_param = new_df.sort_values('acc', ascending=False).iloc[0,1]
```

```python
best_param = ast.literal_eval(best_param)
```
